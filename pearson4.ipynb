```python
import numpy as np
import matplotlib.pyplot as plt
```

## Pearson

```python
def pearson_type_iv_pdf(x, mu, beta, m, nu):
    """
    Compute the PDF of the Pearson Type IV distribution.
    """
    from scipy.special import gamma, loggamma
    # Precompute constants
    const = np.abs(gamma(m + (nu / 2) * 1j))**2 / (gamma(m) * np.sqrt(np.pi * beta))
    term1 = (1 + (x - mu)**2 / beta)**(-m)
    term2 = np.exp(-nu * np.arctan((x - mu) / np.sqrt(beta)))
    
    return const * term1 * term2
```

## Plot

```python
# Parameters for the Pearson Type IV distribution
mu = 0.3    # Location parameter - same as range in Lumerical
m = 6/(7-3) + 2.5    # Shape parameter - approx 6/(kurtosis-3) + 2.5
beta = 0.5**2 * (2*m-3)  # Scale parameter - same is straggle**2*(2m-3)
nu = 0.5*np.sqrt(m)    # Skewness parameter

# Generate x values
x = np.linspace(0, 2.8, 1000)

# Compute the PDF
pdf_values = pearson_type_iv_pdf(x, mu, beta, m, nu)

max_val = np.max(pdf_values)
pdf_values = pdf_values/max_val*2.5e17

# Plot the PDF
# plt.plot(x, pdf_values, label=f"Pearson Type IV (μ={mu}, β={beta}, m={m}, ν={nu})")
plt.figure(figsize=(5, 5))
plt.semilogy(x, pdf_values, label=f"Pearson Type IV (μ={mu}, β={beta}, m={m}, ν={nu})")
plt.xlabel("x")
plt.ylabel("PDF")
plt.title("Pearson Type IV Distribution")
plt.legend()
plt.grid(True)
plt.show()
```
